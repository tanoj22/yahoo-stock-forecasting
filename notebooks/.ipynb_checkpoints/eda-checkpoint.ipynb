```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import mplfinance as mpf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.seasonal import seasonal_decompose

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12,5)
```

```python
df = pd.read_csv("../data/yahoo_stock.csv")
df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values("Date").reset_index(drop=True)
df.head()
```

```python
plt.figure(figsize=(12,5))
plt.plot(df["Date"], df["Close"], label="Close")
plt.title("Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Close")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,5))
for col in ["Open", "High", "Low", "Close"]:
    plt.plot(df["Date"], df[col], label=col, alpha=0.8)
plt.title("OHLC Prices Over Time")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
df_candle = df.set_index("Date")[["Open","High","Low","Close","Volume"]]
mpf.plot(df_candle.tail(150), type="candle", volume=True, mav=(20,50))
```

```python
df["MA20"] = df["Close"].rolling(20).mean()
df["MA50"] = df["Close"].rolling(50).mean()
df["MA100"] = df["Close"].rolling(100).mean()

plt.figure(figsize=(12,6))
plt.plot(df["Date"], df["Close"], label="Close", alpha=0.6)
plt.plot(df["Date"], df["MA20"], label="MA20")
plt.plot(df["Date"], df["MA50"], label="MA50")
plt.plot(df["Date"], df["MA100"], label="MA100")
plt.title("Price with Moving Averages")
plt.legend()
plt.show()
```

```python
df["BB_MID"] = df["Close"].rolling(20).mean()
df["BB_UP"] = df["BB_MID"] + 2*df["Close"].rolling(20).std()
df["BB_LOW"] = df["BB_MID"] - 2*df["Close"].rolling(20).std()

plt.plot(df["Date"], df["Close"], label="Close")
plt.plot(df["Date"], df["BB_MID"], label="Middle Band")
plt.plot(df["Date"], df["BB_UP"], label="Upper Band", linestyle='--')
plt.plot(df["Date"], df["BB_LOW"], label="Lower Band", linestyle='--')
plt.title("Bollinger Bands")
plt.legend()
plt.show()
```

```python
delta = df["Close"].diff()
gain = delta.clip(lower=0)
loss = -delta.clip(upper=0)

avg_gain = gain.rolling(14).mean()
avg_loss = loss.rolling(14).mean()

rs = avg_gain / avg_loss
df["RSI"] = 100 - (100 / (1 + rs))

plt.plot(df["Date"], df["RSI"], color="purple")
plt.axhline(30, color="green", linestyle="--")
plt.axhline(70, color="red", linestyle="--")
plt.title("RSI Indicator")
plt.show()
```

```python
exp1 = df["Close"].ewm(span=12, adjust=False).mean()
exp2 = df["Close"].ewm(span=26, adjust=False).mean()
df["MACD"] = exp1 - exp2
df["Signal"] = df["MACD"].ewm(span=9, adjust=False).mean()

plt.plot(df["Date"], df["MACD"], label="MACD")
plt.plot(df["Date"], df["Signal"], label="Signal")
plt.title("MACD Indicator")
plt.legend()
plt.show()
```

```python
sns.histplot(df["Close"], kde=True)
plt.title("Close Price Distribution")
plt.show()
```

```python
sns.boxplot(x=df["Close"])
plt.title("Close Price  Boxplot")
plt.show()
```

```python
df["Returns"] = df["Close"].pct_change()

sns.histplot(df["Returns"].dropna(), bins=50, kde=True)
plt.title("Daily Returns Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[["Open","High","Low","Close","Adj Close","Volume"]].corr(),
            annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df[["Open","High","Low","Close","Volume"]].sample(300))
plt.show()
```

```python
df["Month"] = df["Date"].dt.month
monthly = df.groupby("Month")["Close"].mean()

monthly.plot(kind="bar", color="teal")
plt.title("Average Close Price by Month")
plt.show()
```

```python
df["Year"] = df["Date"].dt.year
yearly = df.groupby("Year")["Close"].mean()

yearly.plot(marker="o")
plt.title("Average Close Price by Year")
plt.show()
```

```python
df["Day"] = df["Date"].dt.day_name()
dow = df.groupby("Day")["Close"].mean().reindex(
    ["Monday","Tuesday","Wednesday","Thursday","Friday"]
)

dow.plot(kind="bar", color="orange")
plt.title("Average Close Price by Day of Week")
plt.show()
```

```python
df["Volatility"] = df["Returns"].rolling(30).std() * np.sqrt(30)

plt.plot(df["Date"], df["Volatility"])
plt.title("Rolling 30-Day Annualized Volatility")
plt.show()
```

```python
plot_acf(df["Returns"].dropna(), lags=40)
plt.title("Autocorrelation of Returns")
plt.show()
```

```python
summary = pd.DataFrame({
    "Mean Price": [df["Close"].mean()],
    "Volatility (std returns)": [df["Returns"].std()],
    "Max Price": [df["Close"].max()],
    "Min Price": [df["Close"].min()],
    "Best Month": [monthly.idxmax()],
    "Worst Month": [monthly.idxmin()]
})

summary
```
